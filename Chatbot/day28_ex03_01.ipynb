Note  
지금은 RangeIndex이지만 예전 pandas에서는 그냥 정수 인덱스였습니다.  
이를 해결하기 위해 pd.read_csv() 함수의 index_col 파라미터를 사용해   
인덱스로 쓸 컬럼을 지정하면 됩니다.

```python
# 실습-3-해답-ver1-4
df314 = pd.read_csv('data/goog.csv', parse_dates=['Date'], index_col='Date') # parse_dates=['Date'] 속성을 이용해 date 타입으로 변경
df314  # 출력된 Data 열의 모양이 달라진 것을 확인 할 수 있습니다.
```
